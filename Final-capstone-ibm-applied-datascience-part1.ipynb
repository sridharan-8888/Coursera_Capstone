```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="Project_Learning_agent")

def get_geo_location (suburb,city_country_fmt) :
    # initialize your variable to None
    latitude=None
    longitude=None
    location = geolocator.geocode(city_country_fmt.format(suburb))
    if location != None :
        latitude = location.latitude
        longitude = location.longitude
    return latitude, longitude
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

### Explore venues in  Alam Damai

```python
# Explore nearby venues 
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&categoryId={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT,cat_par)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Suburb', 
                  'Suburb Latitude', 
                  'Suburb Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
kl_grouped = kl_onehot_filtered.groupby('Suburb').mean().reset_index()
kl_grouped
```

```python
#Function to sort the venues in descending order.
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#New dataframe and display the top 10 venues for each neighborhood.
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Suburb']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
suburb_venues_sorted = pd.DataFrame(columns=columns)
suburb_venues_sorted['Suburb'] = kl_grouped['Suburb']

for ind in np.arange(kl_grouped.shape[0]):
    suburb_venues_sorted.iloc[ind, 1:] = return_most_common_venues(kl_grouped.iloc[ind, :], num_top_venues)

suburb_venues_sorted.tail()
```

## Cluster Neighborhoods

```python
# set number of clusters
kclusters = 5

kl_grouped_clustering = kl_grouped.drop('Suburb', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(kl_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
suburb_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_.astype('int64'))


kl_sub_df_merged = kl_sub_df

# merge toronto_grouped with kl_data to add latitude/longitude for each suburb
kl_sub_df_merged = kl_sub_df_merged.join(suburb_venues_sorted.set_index('Suburb'), on='Suburb')

kl_sub_df_merged.head() # check the last columns!
kl_sub_df_merged.dropna(axis=0,inplace=True)
```

```python
kl_sub_df_merged.groupby('Cluster Labels').count()
```

```python
latitude = kl_sub_df.loc[0, 'Latitude'] # Suburb latitude value
longitude = kl_sub_df.loc[0, 'Longitude'] # Suburb longitude value

# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(kl_sub_df_merged['Latitude'], kl_sub_df_merged['Longitude'], kl_sub_df_merged['Suburb'], kl_sub_df_merged['Cluster Labels'].astype('int64')):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
